```python
from __future__ import division, print_function

from itertools import izip
import numpy as np
import matplotlib.pyplot as plt
import pylab

from simdb.db import SimulationDatabase
```

```python
config = {'dune_partitioning': '[1 1 1]',
          'dune_num_refinements': 2,
          'dune_linear_solver_options': {'type': 'bicgstab.ilut', 'precision': '1e-14'},
          'mu_bar_value': 0.1,
          'mu_hat_value': 0.1,
          'num_training_samples': 10,
          'greedy_target_error': 1e-14,
          'greedy_use_estimator': True,
          'initialize_with_one': True,
          'extension_product': 'elliptic_plus_penalty',
          'greedy_error_norm': 'elliptic',
          'estimator_return': 'model_reduction_error'}

simdb = SimulationDatabase()
datasets = simdb.select('greedy_OS2014_estimator_study', **config)
dataset = datasets[0]
for ds in datasets:
    if not ds.failed:
        if ds.finished > dataset.finished:
            dataset = ds
print(dataset)
```

```python
estimator_data = dataset.d['estimator_data']
estimator_data.keys()
```

```python
def sorted_the_two(first, second):
    return [list(x) for x in zip(*sorted(zip(first, second), key=lambda pair: pair[0]))]

def get_and_apply(data, key, action=lambda xx: xx):
    tmp_mus =[]
    tmp_values = []
    for mm, vv in izip(data[key + '_mus'], data[key]):
        tmp_mus.append(mm['mu'][0])
        tmp_values.append(action(vv))
    return sorted_the_two(tmp_mus, tmp_values)
```

Let's first take a loot at the discretization errors $\left|\left|\left| u(\mu) - u_h(\mu)\right|\right|\right|_\overline{\mu}$ that we can expect for the $\mu$ in the training set as well as ${\sqrt{\alpha(\mu, \overline{\mu}}}^{-1}$, $\sqrt{\gamma(\mu, \overline{\mu}}$ and ${\sqrt{\alpha(\mu, \hat{\mu}}}^{-1}$.

```python
step = 8
step_data = estimator_data[step]
discretization_errors = get_and_apply(step_data, 'discretization_error')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(discretization_errors[0], discretization_errors[1], 'bD--', label='\'discretization error\': |U(mu) - U_h(mu)|_mu_bar')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('discretization error over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
one_over_sqrt_alpha_bars = get_and_apply(step_data, 'alpha_mu_mu_bar', lambda xx: 1.0/np.sqrt(xx))
sqrt_gamma_bars          = get_and_apply(step_data, 'gamma_mu_mu_bar', lambda xx: np.sqrt(xx))
one_over_sqrt_alpha_hats = get_and_apply(step_data, 'alpha_mu_mu_hat', lambda xx: 1.0/np.sqrt(xx))
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(one_over_sqrt_alpha_bars[0], one_over_sqrt_alpha_bars[1], 'ks--', label='1.0/sqrt(alpha(mu, mu_bar))')
plt.plot(sqrt_gamma_bars[0], sqrt_gamma_bars[1], 'ko--', label='sqrt(gamma(mu, mu_bar))')
plt.plot(one_over_sqrt_alpha_hats[0],one_over_sqrt_alpha_hats[1], 'kD--', label='1.0/sqrt(alpha(mu, mu_hat))')
plt.xlabel('mu')
plt.ylabel('value')
plt.title('parameter change constants over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
step_full_errors = get_and_apply(step_data, 'full_error')
step_model_reduction_errors = get_and_apply(step_data, 'model_reduction_error')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_full_errors[0], step_full_errors[1], 'b+-', label='\'full error\': |U(mu) - U_red(mu)|_mu_bar')
plt.plot(step_model_reduction_errors[0], step_model_reduction_errors[1], 'bo--', label='\'model reduction error\': |U_h(mu) - U_red(mu)|_mu_bar')
plt.plot(discretization_errors[0], discretization_errors[1], 'bD--', label='\'discretization error\': |U(mu) - U_h(mu)|_mu_bar')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('several error over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
step_eta_ncs = get_and_apply(step_data, 'eta_nc_red')
step_eta_rs = get_and_apply(step_data, 'eta_r_red')
step_eta_dfs = get_and_apply(step_data, 'eta_df_red')
step_etas = get_and_apply(step_data, 'eta_red')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_eta_ncs[0], step_eta_ncs[1], 'rv--', label='eta_nc(U_red(mu); mu_bar)')
plt.plot(step_eta_rs[0], step_eta_rs[1], 'r<--', label='eta_r(U_red(mu); mu)')
plt.plot(step_eta_dfs[0], step_eta_dfs[1], 'r>--', label='eta_df(U_red(mu); mu, mu_hat)')
plt.plot(step_etas[0], step_etas[1], 'r', label='\'model reduction error\': eta(U_red(mu); mu, mu_bar, mu_hat)')
#plt.plot(one_over_sqrt_alpha_bars[0], one_over_sqrt_alpha_bars[1], 'ks--', label='1.0/sqrt(alpha(mu, mu_bar))')
#plt.plot(sqrt_gamma_bars[0], sqrt_gamma_bars[1], 'ko--', label='sqrt(gamma(mu, mu_bar))')
#plt.plot(one_over_sqrt_alpha_hats[0], one_over_sqrt_alpha_hats[1], 'kD--', label='1.0/sqrt(alpha(mu, mu_hat))')
plt.xlabel('mu')
plt.title('estimator (components) over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
pylab.rcParams['figure.figsize'] = (20.0, 16.0)
for step in range(0,8):
    step_data = estimator_data[step]
    step_etas = get_and_apply(step_data, 'eta_red')
    discretization_errors = get_and_apply(step_data, 'discretization_error')
    step_full_errors = get_and_apply(step_data, 'full_error')
    step_model_reduction_errors = get_and_apply(step_data, 'model_reduction_error')
    plt.subplot(440+step+1)
    plt.plot(step_full_errors[0], step_full_errors[1], 'b+-', label='full error: |U(mu) - U_red(mu)|_mu_bar')
    plt.plot(step_model_reduction_errors[0], step_model_reduction_errors[1], 'bo--', label='model reduction error: |U_h(mu) - U_red(mu)|_mu_bar')
    plt.plot(step_etas[0], step_etas[1], 'r', label='estimator: eta(U_red(mu); mu, mu_bar, mu_hat)')
    plt.xlabel(' mu ',
               verticalalignment='top',
               horizontalalignment='right')
    plt.ylabel('error')
    plt.title(' step {}'.format(step), horizontalalignment='left')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
