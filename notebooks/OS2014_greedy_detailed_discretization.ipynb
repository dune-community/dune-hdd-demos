```python
from __future__ import division, print_function

from itertools import izip
import numpy as np
import matplotlib.pyplot as plt
import pylab

from simdb.db import SimulationDatabase
```

```python
def extract_datasets(simdb, experiment, config):
    '''
    Extract the experiments of interest. We assume that everything that determines the run
    is specified in config except for `greedy_error_norm` (for the moment assuming that the
    same `extension_product` was used). Of all runs with the same `greedy_error_norm`
    (which should all contain the same results) we take the most recent.
    '''    
    datasets = {}
    for dataset in simdb.select(experiment, **config):
        if not dataset.failed:
            norm = dataset.p.greedy_error_norm            
            if datasets.has_key(norm):
                other_dataset = datasets[norm]
                assert other_dataset.p.extension_product == dataset.p.extension_product
                datasets[norm] = dataset if dataset.finished > other_dataset.finished else other_dataset
            else:
                datasets[norm] = dataset
    return datasets
```

# Investigate the impact of the quality of the detailed discretization

```python
experiment_id = 'greedy-OS2014'
common_config = {'dune_partitioning': '[1 1 1]',
                 'dune_linear_solver_options': {'type': 'bicgstab.ilut', 'precision': '1e-14'},
                 'mu_bar_value': 1.0,
                 'mu_hat_value': 1.0,
                 'initialize_with_one': True,
                 'num_training_samples': 10,
                 'greedy_use_estimator': False,
                 'greedy_target_error': 1e-14}
simdb = SimulationDatabase()
```

## level 1 refinement

```python
level_1_cfg = common_config.copy()
level_1_cfg['dune_num_refinements'] = 1
level_1_datasets = extract_datasets(simdb, experiment_id, level_1_cfg)
for norm, dataset in level_1_datasets.items():
    print('\'{}\': {}'.format(norm, dataset.__repr__()))
```

Let's first take a look at the discretization error

```python
print('max discretization error:')
for norm, dataset in level_1_datasets.items():
    print('  norm {}: {}'.format(norm, np.amax(dataset.d['discretization_errors'])))
```

Can we deduce from this that any solution variation below
* `1` in the `energy` norm
* `1e-3` in the `l2` norm
* `1` in the `elliptic` norm
is not necessaryli due to parameter variation but could also be due to the accuracy of the detailed discretizytion?

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
assert len(level_1_datasets) <= len(colors)
for (norm, dataset), color in izip((level_1_datasets.items()), colors):
    greedy_error_evolution = dataset.d['max_errs']
    plt.plot(greedy_error_evolution, color, label=norm)
plt.yscale('log')
plt.xlabel('extension step')
plt.ylabel('error norm')
plt.title('Error evolution during greedy (level 1 refinement)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## level 2 refinement

```python
level_2_cfg = common_config.copy()
level_2_cfg['dune_num_refinements'] = 2
level_2_datasets = extract_datasets(simdb, experiment_id, level_2_cfg)
purge_norms = []
for norm in level_2_datasets:
    if not level_1_datasets.has_key(norm):
        purge_norms.append(norm)
for norm in purge_norms:
    level_2_datasets.pop(norm)
for norm, dataset in level_2_datasets.items():
    print('\'{}\': {}'.format(norm, dataset.__repr__()))
```

Let's take a look at the discretization errors, compared to level 1

```python
print('discretization error:')
for norm, dataset in level_2_datasets.items():
    print('  norm \'{}\':'.format(norm))
    print('    range: [{} {}]'.format(np.amin(dataset.d['discretization_errors']), np.amax(dataset.d['discretization_errors'])))
```

As expected, the errors are about halfed. We can also take a look at the impact on the greedy error evolution:

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
assert len(level_1_datasets) <= len(colors)
assert len(level_2_datasets) <= len(colors)
for norm, color in izip(level_1_datasets.keys(), colors):
    greedy_error_evolution = level_1_datasets[norm].d['max_errs']
    plt.plot(greedy_error_evolution, color + '--+', label=norm + ' (level 1)')
    greedy_error_evolution = level_2_datasets[norm].d['max_errs']
    plt.plot(greedy_error_evolution, color + '--o', label=norm + ' (level 2)')
plt.yscale('log')
plt.xlabel('extension step')
plt.ylabel('error norm')
plt.title('Error evolution during greedy (for different refinements)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## Level 3 refinement

```python
level_3_cfg = common_config.copy()
level_3_cfg['dune_num_refinements'] = 3
level_3_datasets = extract_datasets(simdb, experiment_id, level_3_cfg)
for norm, dataset in level_3_datasets.items():
    print('\'{}\': {}'.format(norm, dataset.__repr__()))
```

```python
print('max discretization error:')
for norm, dataset in level_3_datasets.items():
    print('  norm {}: {}'.format(norm, np.amax(dataset.d['discretization_errors'])))
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for norm, color in izip(level_1_datasets.keys(), colors):
    greedy_error_evolution = level_1_datasets[norm].d['max_errs']
    plt.plot(greedy_error_evolution, color + '--+', label=norm + ' (level 1)')
    greedy_error_evolution = level_2_datasets[norm].d['max_errs']
    plt.plot(greedy_error_evolution, color + '--o', label=norm + ' (level 2)')
    greedy_error_evolution = level_3_datasets[norm].d['max_errs']
    plt.plot(greedy_error_evolution, color + '-s', label=norm + ' (level 3)')
plt.yscale('log')
plt.xlabel('extension step')
plt.ylabel('error norm')
plt.title('Error evolution during greedy (for different refinements)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
