```python
from __future__ import division, print_function

from itertools import izip
import numpy as np
import matplotlib.pyplot as plt
import pylab

from simdb.db import SimulationDatabase
```

```python
config = {'dune_partitioning': '[1 1 1]',
          'dune_num_refinements': 2,
          'dune_linear_solver_options': {'type': 'bicgstab.ilut', 'precision': '1e-14'},
          'mu_bar_value': 0.1,
          'mu_hat_value': 0.1,
          'num_training_samples': 10,
          'greedy_target_error': 1e-14,
          'greedy_use_estimator': True,
          'initialize_with_one': True,
          'extension_product': 'elliptic_plus_penalty',
          'greedy_error_norm': 'elliptic'}

simdb = SimulationDatabase()
datasets = simdb.select('greedy_OS2014_estimator_study', **config)
dataset = datasets[0]
for ds in datasets:
    if not ds.failed:
        if ds.finished > dataset.finished:
            dataset = ds
print(dataset)
```

```python
estimator_data = dataset.d['estimator_data']
estimator_data.keys()
```

```python
def sorted_the_two(first, second):
    return [list(x) for x in zip(*sorted(zip(first, second), key=lambda pair: pair[0]))]

def get_and_apply(data, key, action=lambda xx: xx):
    tmp_mus =[]
    tmp_values = []
    for mm, vv in izip(data[key + '_mus'], data[key]):
        tmp_mus.append(mm['mu'][0])
        tmp_values.append(action(vv))
    return sorted_the_two(tmp_mus, tmp_values)
```

Let's first take a loot at the discretization errors $\left|\left|\left| u(\mu) - u_h(\mu)\right|\right|\right|_\overline{\mu}$ that we can expect for the $\mu$ in the training set as well as ${\sqrt{\alpha(\mu, \overline{\mu}}}^{-1}$, $\sqrt{\gamma(\mu, \overline{\mu}}$ and ${\sqrt{\alpha(\mu, \hat{\mu}}}^{-1}$.

```python
step_0_data = estimator_data[0]
discretizaiton_errors = get_and_apply(step_0_data, 'discretization_error')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_0_discretizaiton_errors[0], step_0_discretizaiton_errors[1], 'bD--', label='\'discretization error\': |U(mu) - U_h(mu)|_mu_bar')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('discretization error over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
one_over_sqrt_alpha_bars = get_and_apply(step_0_data, 'alpha_mu_mu_bar', lambda xx: 1.0/np.sqrt(xx))
sqrt_gamma_bars          = get_and_apply(step_0_data, 'gamma_mu_mu_bar', lambda xx: np.sqrt(xx))
one_over_sqrt_alpha_hats = get_and_apply(step_0_data, 'alpha_mu_mu_hat', lambda xx: 1.0/np.sqrt(xx))
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(one_over_sqrt_alpha_bars[0], one_over_sqrt_alpha_bars[1], 'ks--', label='1.0/sqrt(alpha(mu, mu_bar))')
plt.plot(sqrt_gamma_bars[0], sqrt_gamma_bars[1], 'ko--', label='sqrt(gamma(mu, mu_bar))')
plt.plot(one_over_sqrt_alpha_hats[0],one_over_sqrt_alpha_hats[1], 'kD--', label='1.0/sqrt(alpha(mu, mu_hat))')
plt.xlabel('mu')
plt.ylabel('value')
plt.title('parameter change constants over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# initial reduced basis

Let us now take a look at the full error of the reduced solution $\left|\left|\left| u(\mu) - u_\text{red}(\mu)\right|\right|\right|_\overline{\mu}$ and the model reduction error $\left|\left|\left| u_h(\mu) - u_\text{red}(\mu)\right|\right|\right|_\overline{\mu}$.

```python
step_0_full_errors = get_and_apply(step_0_data, 'full_error')
step_0_model_reduction_errors = get_and_apply(step_0_data, 'model_reduction_error')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_0_full_errors[0], step_0_full_errors[1], 'b+-', label='\'full error\': |U(mu) - U_red(mu)|_mu_bar')
plt.plot(step_0_model_reduction_errors[0], step_0_model_reduction_errors[1], 'bo--', label='\'model reduction error\': |U_h(mu) - U_red(mu)|_mu_bar')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('several error over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

As we can see (for the initial basis) the model reduction error is very close to the full error, meaning that the reduced solution is as far away from the discrete solution as it is from the weak solution.

Let us now have a look at some estimator components.

```python
step_0_eta_ncs = get_and_apply(step_0_data, 'eta_nc_red')
step_0_eta_rs = get_and_apply(step_0_data, 'eta_r_red')
step_0_eta_dfs = get_and_apply(step_0_data, 'eta_df_red')
step_0_etas = get_and_apply(step_0_data, 'eta_red')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_0_eta_ncs[0], step_0_eta_ncs[1], 'rv--', label='eta_nc(U_red(mu); mu_bar)')
plt.plot(step_0_eta_rs[0], step_0_eta_rs[1], 'r<--', label='eta_r(U_red(mu); mu)')
plt.plot(step_0_eta_dfs[0], step_0_eta_dfs[1], 'r>--', label='eta_df(U_red(mu); mu, mu_hat)')
plt.plot(step_0_etas[0], step_0_etas[1], 'r', label='\'model reduction error\': eta(U_red(mu); mu, mu_bar, mu_hat)')
plt.plot(one_over_sqrt_alpha_bars[0], one_over_sqrt_alpha_bars[1], 'ks--', label='1.0/sqrt(alpha(mu, mu_bar))')
plt.plot(sqrt_gamma_bars[0], sqrt_gamma_bars[1], 'ko--', label='sqrt(gamma(mu, mu_bar))')
plt.plot(one_over_sqrt_alpha_hats[0], one_over_sqrt_alpha_hats[1], 'kD--', label='1.0/sqrt(alpha(mu, mu_hat))')
plt.xlabel('mu')
plt.title('estimator (components) over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

We can observe that the estimator component $\eta_\text{r}$ has the highest contribution to the estimator, which is not surprising since the initial reduced basis yields reduced solutions, the divergence of which is a long way from $\Pi_0(f)$.

Now comparing the estimator to the full error and the model reduction error we can observe that the estimator is an upper bound to the full error.

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_0_full_errors[0], step_0_full_errors[1], 'b+-', label='full error: |U(mu) - U_red(mu)|_mu_bar')
plt.plot(step_0_model_reduction_errors[0], step_0_model_reduction_errors[1], 'bo--', label='model reduction error: |U_h(mu) - U_red(mu)|_mu_bar')
plt.plot(step_0_etas[0], step_0_etas[1], 'r', label='estimator: eta(U_red(mu); mu, mu_bar, mu_hat)')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('several errors over the training set')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The greeds adds a snapshot for the lowest parameter $\mu_0$ to the basis:

```python
print(dataset.d['max_err_mus'][0])
```

```python
step_1_data = estimator_data[1]
step_1_full_errors = get_and_apply(step_1_data, 'full_error')
step_1_model_reduction_errors = get_and_apply(step_1_data, 'model_reduction_error')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(discretizaiton_errors[0], discretizaiton_errors[1], 'bD--', label='\'discretization error\': |U(mu) - U_h(mu)|_mu_bar')
plt.plot(step_1_full_errors[0], step_1_full_errors[1], 'b+-', label='\'full error\': |U(mu) - U_red(mu)|_mu_bar')
plt.plot(step_1_model_reduction_errors[0], step_1_model_reduction_errors[1], 'bo--', label='\'model reduction error\': |U_h(mu) - U_red(mu)|_mu_bar')
plt.xlabel('mu')
plt.ylabel('error')
plt.title('several errors over the training set after the first extension')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
step_1_eta_ncs = get_and_apply(step_1_data, 'eta_nc_red')
step_1_eta_rs = get_and_apply(step_1_data, 'eta_r_red')
step_1_eta_dfs = get_and_apply(step_1_data, 'eta_df_red')
step_1_etas = get_and_apply(step_1_data, 'eta_red')
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(step_1_eta_ncs[0], step_1_eta_ncs[1], 'rv--', label='eta_nc(U_red(mu); mu_bar)')
plt.plot(step_1_eta_rs[0], step_1_eta_rs[1], 'r<--', label='eta_r(U_red(mu); mu)')
plt.plot(step_1_eta_dfs[0], step_1_eta_dfs[1], 'r>--', label='eta_df(U_red(mu); mu, mu_hat)')
plt.plot(step_1_etas[0], step_1_etas[1], 'r', label='eta(U_red(mu); mu, mu_bar, mu_hat)')
#plt.plot(one_over_sqrt_alpha_bars[0], one_over_sqrt_alpha_bars[1], 'ks--', label='1.0/sqrt(alpha(mu, mu_bar))')
#plt.plot(sqrt_gamma_bars[0], sqrt_gamma_bars[1], 'ko--', label='sqrt(gamma(mu, mu_bar))')
#plt.plot(one_over_sqrt_alpha_hats[0], one_over_sqrt_alpha_hats[1], 'kD--', label='1.0/sqrt(alpha(mu, mu_hat))')
plt.xlabel('mu')
plt.title('estimator (components) over the training set after 1 extension')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
