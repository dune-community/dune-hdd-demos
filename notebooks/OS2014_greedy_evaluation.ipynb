```python
from __future__ import division, print_function

from itertools import izip
import numpy as np
import matplotlib.pyplot as plt
::import pylab

from simdb.db import SimulationDatabase
```

```python
simdb = SimulationDatabase()

def extract_datasets(experiment, config):
    '''
    Extract the experiments of interest. We assume that everything that determines the run
    is specified in config except for `greedy_error_norm` (for the moment assuming that the
    same `extension_product` was used). Of all runs with the same `greedy_error_norm`
    (which should all contain the same results) we take the most recent.
    '''    
    datasets = {}
    for dataset in simdb.select(experiment, **config):
        if not dataset.failed:
            norm = dataset.p.greedy_error_norm
            assert dataset.p.extension_product == norm
            if datasets.has_key(norm):
                other_dataset = datasets[norm]
                datasets[norm] = dataset if dataset.finished > other_dataset.finished else other_dataset
            else:
                datasets[norm] = dataset
    return datasets
```

# Investigate the role of the initial basis

We investigate the impact of the initial basis, i.e. if we start with an empty basis or with the `1` (as is needed for the estimator).

```python
experiment_id = 'greedy-OS2014'
common_config = {'dune_partitioning': '[1 1 1]',
                 'dune_num_refinements': 2,
                 'dune_linear_solver_precision': '1e-14',
                 'mu_bar_value': 1.0,
                 'num_training_samples': 100,
                 'greedy_target_error': 1e-12}
```

## empty initial basis

```python
without_one_cfg = common_config.copy()
without_one_cfg['initialize_with_one'] = False
without_one_datasets = extract_datasets(experiment_id, without_one_cfg)
for norm, dataset in without_one_datasets.items():
    print('\'{}\': {}'.format(norm, dataset.__repr__()))
```

We can take a look at these datasets

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
assert len(without_one_datasets) <= len(colors)
for (norm, dataset), color in izip((without_one_datasets.items()), colors):
    greedy_error_evolution = dataset.d['max_errs']
    plt.plot(greedy_error_evolution, color, label=norm)
plt.yscale('log')
plt.xlabel('extension step')
plt.ylabel('error norm')
plt.title('Error evolution during greedy (empty initial basis)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

We can also access the logs of each run to see the reason why the greedy exited, for instance.

```python
for norm, dataset in without_one_datasets.items():
    print('Log for norm \'{}\':'.format(norm))
    print(80*'=')
    print(dataset.d['log'])
```

## initial basis with 1

```python
with_one_cfg = common_config.copy()
with_one_cfg['initialize_with_one'] = True
with_one_datasets = extract_datasets(experiment_id, with_one_cfg)
for norm, dataset in with_one_datasets.items():
    print('\'{}\': {}'.format(norm, dataset.__repr__()))
```

Now we can compare the error evolution during greedy with and without the `1` in the initial basis.

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
assert len(without_one_datasets) <= len(colors)
assert len(with_one_datasets) <= len(colors)
for (norm, dataset), color in izip((without_one_datasets.items()), colors):
    greedy_error_evolution = dataset.d['max_errs']
    plt.plot(greedy_error_evolution, color, label=norm)
for (norm, dataset), color in izip((with_one_datasets.items()), colors):
    greedy_error_evolution = dataset.d['max_errs']
    plt.plot(greedy_error_evolution, color + '--o', label=norm + ' (with 1)')
plt.yscale('log')
plt.xlabel('extension step')
plt.ylabel('error norm')
plt.title('Error evolution during greedy (for different initial bases)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

We can also compare the logs for the different runs, i.e. what the greedy did in the `l2_plus_h1_semi` (which is `H1`) case.

```python
for norm, log in [('l2_plus_h1_semi (without one)', without_one_datasets['l2_plus_h1_semi'].d['log']),
                  ('l2_plus_h1_semi (with one)',    with_one_datasets['l2_plus_h1_semi'].d['log'])]:
    print('Log for norm \'{}\':'.format(norm))
    print(80*'=')
    print(log)
```

We can also take a closer look at the `H1` case and check if the max error is smaller if we incorporate the `1`.

```python
max_errs_h1_without_one = without_one_datasets['l2_plus_h1_semi'].d['max_errs']
max_errs_h1_with_one    = with_one_datasets['l2_plus_h1_semi'].d['max_errs']
for err_without, err_with in izip(max_errs_h1_without_one, max_errs_h1_with_one):
    print('{} vs. {} ({})'.format(err_without, err_with, 'smaller' if err_with < err_without else 'larger'))
```
